```python
import pandas as pd
```

```python
dados = pd.read_json('/content/dados_hospedagem.json')
dados.head()
```

Podemos perceber que nossos dados estão aninhados e dicionários. Pra termos um estudo desses dados, prepará-os e manipula-los, eles precisam estar destrinchados. Então precisamos aplicar a "normalização" e para isso utilizaremos o método `json_normalize`.

O método json_normalize() é usado para normalizar um objeto JSON que pode ter uma ou várias hierarquias. Quando o objeto é transformado para um formato de tabela com múltiplas colunas, cada coluna representa uma chave do JSON e cada linha representa um registro. É uma forma de tornar os dados mais fáceis de manipular e analisar.

```python
dados = pd.json_normalize(dados['info_moveis']) #Na [] coloca-se o nome da coluna
dados
```

Em algumas colunas, como quantidade de quartos, modelo de cama, comodidades, etc, temos dados agrupados em listas. Para facilitar os acesso aos dados, precisamos desagrupá-los. Para isso utilizaremos o método `explode`

```python
#O primeiro passo é coletar quais são os nomes das nossas colunas
colunas = list(dados.columns)
colunas
```

```python
#Vamos manter a variável "dados", pois queremos reatribuí-la para um DataFrame que contenha os dados desagrupados
dados = dados.explode(colunas[3:]) #Usamos o índice 3 pois é a partir da coluna 'descricao_local' até o final que estão agrupados
dados
```

Podemos notar agora que os índices não estão em ordem. Temos repetidos varias vezes, por exemplo, o índice zero. Portanto, precisamos resetá-lo através do método `reset_index`

```python
dados.reset_index(inplace = True, drop = True)
dados.head()
# inplace = True : Transformação deve substituir o DataFrame
# drop = True : Quando damos apenas o reset_index, ele cria uma coluna nova com o índice resetado para "preservar" o índice antigo. Como não precisamos do índice antigo, o drop = True elimina essa nova coluna
```

```python
#Método info retorna os tipos de dados que temos
dados.info()
```

Conseguimos observar que todas as colunas possuem dados do tipo "object". Isso pode representar três cenários:

*   A mesma coluna possui vários tipos diferentes (string, float, etc) de dados e o Panda não consegue devolver essa informação
*   A coluna possui um tipo de arquivo específico, mas esse tipo não foi identificado
*   Se todos os dados forem string

Quando temos várias colunas do tipo object, a transformação para tratar esses dados e manipula-los precisa ser feita manualmente.



Começaremos essa transformação pelo dados numéricos, para isso, usaremos o método `.astype`. Pra utilizar esse método, precisamos importar a biblioteca `numpy`, que nos fornece tipos de dados (int, float) para que possamos converter nossas colunas numéricas.

Vamos começar nossa conversão pela coluna max_hospedes

```python
import numpy as np
```

```python
dados['max_hospedes'] = dados['max_hospedes'].astype(np.int64)
```

```python
dados.info()
#Observamos que a coluna max_hospedes já foi atualizada para o tipo int64
```

```python
col_numericas = ['quantidade_banheiros','quantidade_quartos','quantidade_camas']
```

```python
dados[col_numericas] = dados[col_numericas].astype(np.int64)
```

```python
dados['avaliacao_geral'] = dados['avaliacao_geral'].astype(np.float64)
```

```python
dados.info()
```

```python
dados['preco'] = dados['preco'].astype(np.float64) #Vai dar erro, pois o programa vê o $ e já não entende como converter
```

```python
dados['preco']
```

Podemos observar que cada elemento da coluna 'preco' é uma string. Se fizermos uma função que roda elemento a elemento, linha por linha da coluna 'preco' e converter a string para remover esses caracteres ($ e .), podemos tratar cada um desses dados para serem transformados em float. Faremos isso através do método `lambda`

```python
dados['preco'].apply(lambda x: x.replace('$', '').replace(',','').strip())
# lambda : lê os elementos (chamamos o elemento de x)
# x.replace : indica os elementos que devem ser substituidos e por quais devem ser substituidos (como indicamos '' vazias, os elementos serão apenas apagados.)
```

```python
dados['preco'] = dados['preco'].apply(lambda x: x.replace('$', '').replace(',','').strip())
```

```python
dados['preco'] = dados['preco'].astype(np.float64)
```

```python
dados.info()
```

Para realizar a conversão de múltiplas colunas simultaneamente utilizaremos o método `applymap`

```python
dados[['taxa_deposito','taxa_limpeza']] = dados[['taxa_deposito','taxa_limpeza']].applymap(lambda x: float(x.replace('$', '').replace(',','').strip()))
dados[['taxa_deposito','taxa_limpeza']]
```

```python
dados[['taxa_deposito','taxa_limpeza']] = dados[['taxa_deposito','taxa_limpeza']].astype(np.float64)
```

```python
dados.info()
```

Para realizar um estudo da parte textual dos dados é interessante utilizar uma estrutura de tokenização simples, que significa segregar o texto em unidades menores, "tokens" (palavras, sílabas, etc)

```python
dados['descricao_local'] = dados['descricao_local'].str.lower()
#.str Faz a leitura dos elementos como strings
#lower() Converte todas as letras para minúscula
```

```python
dados.head()
```

Para remover caracteres que não trazem informação (/, ?, ! ", etc), podemos utilizar o comando .str ou a expressão regular Regex, que se caracteriza como o conjunto de caracteres que servem para fazer uma pesquisa no texto. A partir dessa sequência padronizada de caracteres, podemos definir o que queremos ou não selecionar.

```python
dados['descricao_local'][3169]
```

```python
dados['descricao_local'].str.replace('[^a-zA-Z0-9\-\']', ' ', regex=True)
# [^a-zA-Z0-9\-\'] é o Regex, é o conjunto de caracteres que pesquisaremos no texto
# O ^ define os caracteres que não serão selecionados. Então todos os caracteres, com exceção desses, serão selecionados
```

```python
dados['descricao_local'] = dados['descricao_local'].str.replace('[^a-zA-Z0-9\-\']', ' ', regex=True)
```

```python
dados['descricao_local'].str.replace('(?<!\w)-(?!\w)', ' ', regex=True)
# O Regex (?<!\w)-(?!\w) serve para verificar se existem caracteres antes e depois do hífen (-)
```

```python
dados['descricao_local'] = dados['descricao_local'].str.replace('(?<!\w)-(?!\w)', ' ', regex=True)
```

Para criar os tokens precisaremos separar cada palavra do texto como elementos de uma lista. Para isso, utilizaremos o método `split()`

```python
dados['descricao_local'].str.split()
```

```python
dados['descricao_local'] = dados['descricao_local'].str.split()
dados.head()
```

```python
dados['comodidades'] = dados['comodidades'].str.replace('\{|}|\"','',regex=True)
dados.head()
```

Dentro da coluna "comodidades", os itens estão separados por vírgulas. Após utilizar o Regex para remover os caracters indesejados ({, ", etc), nós aproveitaremos as vírgulas como separadores, indicando no split que a divisão seja feita nelas

```python
dados['comodidades'] = dados['comodidades'].str.split(',')
dados.head()
```

Trabalharemos agora com uma segunda base de dados (`moveis_disponiveis.json`.) que contém as vagas diárias disponibilizadas em vários imóveis para hospedagem durante o ano de 2016. Esta base mostra a oferta de imóveis disponíveis para aluguel durante este ano, o que pode ser interessante para a precificação inteligente, o foco da nossa aplicação.

Isso quer dizer que lidaremos com dados de tempo, ou seja, o tipo de dado `datetime`.

A classe `datetime` é um tipo de dado que representa uma data e hora específicas em Python. Ela permite trabalhar e manipular dados de ano, mês, dia, hora, minuto, segundo e microssegundo, além de dias da semana, como segunda, terça, quarta, etc. Essa classe é definida no módulo `datetime` e é fundamental na manipulação de datas e horários em Python.



```python
dt_data = pd.read_json('/content/moveis_disponiveis.json')
dt_data.head()
```

O `datetime` possui um formato padrão de apresentação da data: ANO - MÊS - DIA

```python
dt_data.info()
```

Podemos observar que o tipo da coluna data está em formato "object", ou seja, precisamos transformá-lo para o tipo "datetime" através do método `to_datetime`

```python
pd.to_datetime(dt_data['data'])
```

```python
dt_data['data'] = pd.to_datetime(dt_data['data'])
```

```python
dt_data.info()
```

```python
dt_data.head()
```

```python
dt_data['data'].dt.strftime('%Y-%m')
#srtftime é capaz de formatar os valores de data para o formato que quisermos
```

Agora que transformamos as datas somente para ano e mês através do método `strftime`, conseguiremos agrupar todos esses meses e as vagas disponíveis em cada mês, utilizarando o método `groupby`.

```python
subset = dt_data.groupby(dt_data['data'].dt.strftime('%Y-%m'))['vaga_disponivel'].sum()
subset
#(dt_data['data'].dt.strftime('%Y-%m')) é a categoria que será agrupada
#['vaga_disponivel'] é a a coluna a ser agrupada por mês
```
